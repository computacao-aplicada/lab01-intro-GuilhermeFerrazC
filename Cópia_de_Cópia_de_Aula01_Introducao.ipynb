## O que √© Vis√£o Computacional?

A **Vis√£o Computacional** √© uma √°rea da Intelig√™ncia Artificial que busca permitir que os computadores 'vejam' e interpretem o mundo visual, assim como os humanos.

Ela utiliza imagens e v√≠deos como entrada e tenta extrair informa√ß√µes √∫teis, como:
- Detec√ß√£o de objetos
- Reconhecimento facial
- Segmenta√ß√£o de imagem
- Rastreamento de movimento

A **Computa√ß√£o Gr√°fica**, por outro lado, est√° mais voltada √† gera√ß√£o de imagens ‚Äî como renderizar cenas em 3D.

J√° o **Processamento de Imagens** se concentra na transforma√ß√£o e an√°lise de imagens (normalmente 2D) para melhorar a qualidade ou extrair caracter√≠sticas.

## Exemplo pr√°tico: Carregar e visualizar uma imagem

```python
# Importar bibliotecas
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import requests

# Carregar imagem
url = "https://images.unsplash.com/photo-1745810187217-4d9e1ccfd9d5?ixlib=rb-4.1.0&q=85&fm=jpg&crop=entropy&cs=srgb&w=640"
image = Image.open(requests.get(url, stream=True).raw).convert("RGB")

# Mostrar imagem original
plt.imshow(image)
plt.title("Imagem Original")
plt.axis('off')
plt.show()
```

## Manipula√ß√£o dos canais RGB

```python
# Separar canais RGB
r, g, b = image.split()

# Criar imagem apenas com o canal vermelho
red_channel = Image.merge("RGB", (r, Image.new("L", image.size), Image.new("L", image.size)))

# Exibir
plt.imshow(red_channel)
plt.title("Apenas Canal Vermelho")
plt.axis('off')
plt.show()
```

```python
# Converter para array numpy
np_image = np.array(image)

# Modificar canal vermelho (R) com valores aleat√≥rios
np_image[:, :, 0] = np.random.randint(0, 256, (np_image.shape[0], np_image.shape[1]), dtype=np.uint8)

# Converter de volta para imagem
modified_image = Image.fromarray(np_image)

# Exibir
plt.imshow(modified_image)
plt.title("Canal Vermelho Modificado")
plt.axis('off')
plt.show()
```

```python
# Inverter cores (255 - valor do pixel)
negative_image = Image.fromarray(255 - np.array(image))

# Exibir
plt.imshow(negative_image)
plt.title("Imagem Negativa")
plt.axis('off')
plt.show()
```

```python
negative_image.save("imagem_negativa.jpg")
print("imagem_negativa.jpg")
```

```python
# Coordenadas (left, top, right, bottom)
crop_coords = (150, 100, 450, 400)  # Ajuste conforme o tamanho da sua imagem

# Recortar
cropped_image = image.crop(crop_coords)

# Exibir
plt.imshow(cropped_image)
plt.title("Recorte da Imagem Original")
plt.axis('off')
plt.show()

# Salvar
cropped_image.save("recorte_original.jpg")
print("recorte_original.jpg")
```

```python
# Usar a imagem negativa criada anteriormente
cropped_negative = negative_image.crop(crop_coords)  # Mesmas coordenadas

# Exibir
plt.imshow(cropped_negative)
plt.title("Recorte da Imagem Negativa")
plt.axis('off')
plt.show()

# Salvar
cropped_negative.save("recorte_negativo.jpg")
print("recorte_negativo.jpg")
```
